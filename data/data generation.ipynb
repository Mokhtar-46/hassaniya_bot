```python

# ----------------------------------------------------------------
# FONCTIONS DE GÉNÉRATION
# ----------------------------------------------------------------

def remplir(template, cat):
    """Remplace les slots lexicaux dans un template."""
    s = template
    s = s.replace("{prenom}",        random.choice(prenoms_masculins))
    s = s.replace("{prenom_f}",      random.choice(prenoms_feminins))
    s = s.replace("{lieu}",          random.choice(lieux))
    s = s.replace("{marchandise}",   random.choice(marchandises))
    s = s.replace("{prix}",          random.choice(prix))
    s = s.replace("{moment}",        random.choice(moments_journee))
    s = s.replace("{maladie}",       random.choice(maladies))
    s = s.replace("{remede}",        random.choice(remedes))
    s = s.replace("{saison}",        random.choice(saisons))
    s = s.replace("{meteo}",         random.choice(meteos))
    s = s.replace("{conseil_sujet}", random.choice(conseils_sujets))
    s = s.replace("{ouv}",           random.choice(ouvertures))
    s = s.replace("{fer}",           random.choice(fermetures))
    return s

def generer_paire_template(cat):
    """Génère une paire Q/R depuis les templates."""
    data = templates[cat]
    q = remplir(random.choice(data["q_templates"]), cat)
    r = remplir(random.choice(data["r_templates"]), cat)
    return {"input": q, "output": r, "categorie": cat}

def generer_paire_seed(cat):
    """Utilise une graine existante avec variation légère."""
    data = templates[cat]
    q, r = random.choice(data["seeds"])
    # Variation : ajouter une ouverture ou fermeture aléatoire
    if random.random() > 0.5 and not r.startswith("و"):
        r = random.choice(ouvertures) + r
    if random.random() > 0.6 and not r.endswith("الله") and not r.endswith("نا"):
        r = r + random.choice(fermetures)
    # Remplir les slots si présents
    q = remplir(q, cat)
    r = remplir(r, cat)
    return {"input": q, "output": r, "categorie": cat}

def generer_paire_hybride(cat1, cat2):
    """Mélange contrôlé : question d'une catégorie, intro d'une autre (stratégie 6)."""
    data1 = templates[cat1]
    data2 = templates[cat2]
    q = remplir(random.choice(data1["q_templates"]), cat1)
    # Réponse hybride : intro cat2 + réponse cat1
    r_base = remplir(random.choice(data1["r_templates"]), cat1)
    intro = remplir(random.choice(data2["r_templates"]), cat2)
    r = intro.split("،")[0] + "، " + r_base if "،" in intro else r_base
    return {"input": q, "output": r, "categorie": cat1}

def appliquer_negation(paire):
    """Variation de polarité : transforme une affirmation en question négative."""
    q = paire["input"]
    neg = random.choice(negations_p)
    if q.startswith(" شنه "):
        q_neg = neg + q[4:]
    else:
        q_neg = "شنه " + neg + q
    return {"input": q_neg, "output": paire["output"], "categorie": paire["categorie"]}

# ----------------------------------------------------------------
# ASSEMBLAGE DU DATASET
# ----------------------------------------------------------------

categories = list(templates.keys())
dataset    = []

TARGET = 10000
par_cat = TARGET // len(categories)  # ~1000 par catégorie

for cat in categories:
    count = 0

    # 1. Seeds (15%)
    while count < int(par_cat * 0.15):
        dataset.append(generer_paire_seed(cat))
        count += 1

    # 2. Templates purs (55%)
    while count < int(par_cat * 0.70):
        dataset.append(generer_paire_template(cat))
        count += 1

    # 3. Variation de négation (15%) - only for prix
    while count < int(par_cat * 0.85):
        if cat == "prix":
            paire = generer_paire_template(cat)
            dataset.append(appliquer_negation(paire))
        else:
            dataset.append(generer_paire_template(cat))
        count += 1

    # 4. Hybride avec catégorie voisine (15%)
    cat_voisine = random.choice([c for c in categories if c != cat])
    while count < par_cat:
        dataset.append(generer_paire_hybride(cat, cat_voisine))
        count += 1

# Compléter jusqu'à exactement 10000+
while len(dataset) < TARGET:
    cat = random.choice(categories)
    dataset.append(generer_paire_template(cat))

print(f"Total généré : {len(dataset)} paires")
print(f"Répartition : { {c: sum(1 for d in dataset if d['categorie']==c) for c in categories} }")
```

```python

# ----------------------------------------------------------------
# SPLIT TRAIN / TEST (95 / 5)
# ----------------------------------------------------------------
random.shuffle(dataset)
split = int(len(dataset) * 0.95)
train_data, test_data = dataset[:split], dataset[split:]

with open("train.json", "w", encoding="utf-8") as f:
    json.dump(train_data, f, ensure_ascii=False, indent=2)

with open("test.json", "w", encoding="utf-8") as f:
    json.dump(test_data, f, ensure_ascii=False, indent=2)

print(f"\nTrain : {len(train_data)} | Test : {len(test_data)}")
print("Fichiers sauvegardés : train.json, test.json")
```
